```python
def fact(n):
    if n==0: return 1
    return n * fact(n-1)
```

```python
fact(30)
```

```python
F = {}
def fib(n):
    if n==0 or n==1: return 1
    if n in F: return F[n]
    F[n] = fib(n-1) + fib(n-2)    
    return F[n]
```

```python
for i in range(10):
    print(fib(i))
```

```python
class TreeNode:
    def __init__(self, v):
        self.v = v
        self.left = self.right = None

    def contains(self, v):
        if v==self.v:
            return True
        if v<self.v and self.left:
            return self.left.contains(v)
        if v>self.v and self.right:
            return self.right.contains(v)
        return False
        
```

```python
from lolviz import *
root = TreeNode(2)
root.left = TreeNode(1)
root.right = TreeNode(3)
treeviz(root)
```

```python
for i in range(0,5):
    print(i, root.contains(i))
```

```python
def add(p, v):
    if p is None:
        return TreeNode(v)
    if v==p.v:
        return p
    if v<p.v:
        p.left = add(p.left, v)
        return p
    if v>p.v:
        p.right = add(p.right, v)
        return p
```

```python
root = TreeNode(10)
root = add(root, 5)
root = add(root, 30)
root = add(root, 2)
root = add(root, 8)
treeviz(root)
```
